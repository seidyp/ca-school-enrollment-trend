```python
# dependencies & set up
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pathlib
from IPython.display import display

# data csv includes unemployment data per year
econ_filepath = "Resources/CensusDataFinal.csv"

# district csv includes enrollment totals per district per year along with percent changes
district_filepath = "Resources/output_data/enrollment_by_dist_changes.csv"

# school directory
school_filepath = "Resources/ca-school-list.csv"

# converting csvs into usable dfs
econ_df = pd.read_csv(econ_filepath)
district_df = pd.read_csv(district_filepath)
school_df = pd.read_csv(school_filepath)
```

```python
# verifying data types of dfs to be merged - confirming data type of common column is matching 
econ_df.dtypes

district_df.dtypes
```

```python
# merging economic and district enrollment data based on district code and year
econ_district_df = pd.merge(district_df, econ_df, on=["NCESDist", "year"])
```

```python
econ_district_df.columns
```

```python
# dropping old index columns
econ_district_df = econ_district_df.drop(columns=["Unnamed: 0_x", "Unnamed: 0_y"])
```

```python
#verifying data types again 
econ_district_df.dtypes
```

```python
# cleaning up column names to make it more readable
econ_district_df = econ_district_df.rename(columns={
    "year": "Year", 
    "ENR_TOTAL": "Total Enrollment", 
    "Total Population, 16+": "Population, 16+", 
    "Employment/Population Ratio!!Population 16 years and over":"Employment Ratio, 16+", 
    "Unemployment Rate. 16+":"Unemployment Rate, 16+", 
    "Total!!Population 20 to 64 years":"Population, 20-64", 
    "Employment/Population Ratio!!Population 20 to 64 years":"Employment Ratio, 20-64", 
    "Unemployment rate!!Population 20 to 64 years":"Unemployment Rate, 20-64"})
```

```python
econ_district_df = econ_district_df.replace(-999999999.0,np.nan)
econ_district_df
```

```python
# exporting to csv 
econ_district_df.to_csv("Resources/output_data/econ_district_merged.csv")
```
