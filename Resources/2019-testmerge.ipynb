```python
econ_data_df = pd.read_csv(econ_data_path)
school_data19_df = pd.read_csv(school_enrollment_path_19)
```

```python
# data types must match for them to be joined on
school_data19_df.dtypes
```

```python
is_2019 = school_data19_df["year"] == 2019
school19_df = school_data19_df.loc[is_2019, :]
```

```python
district_groups_df = school19_df.groupby(by=["NCESDist"]).sum()
district_enroll_total_df = district_groups_df["ENR_TOTAL"].reset_index()
district_enroll_total_df
```

```python
new_econ_df = econ_data_df.astype({"NCESDist":"object"})
new_district_df = district_enroll_total_df["ENR_TOTAL"].astype({"ENR_TOTAL":"object"})
```
