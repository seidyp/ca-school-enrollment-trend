```python
# Dependencies and Setup
import pandas as pd
import pathlib
import csv
import scipy.stats as st
import numpy as np
```

```python
# Create paths for csv

employment_unified_2018_data = "Merged Employment/Resources/output_employment_data/Unified2018.csv"
employment_elementary_2018_data = "Merged Employment/Resources/output_employment_data/Elementary2018.csv"
employment_secondary_2018_data = "Merged Employment/Resources/output_employment_data/Secondary2018.csv"
```

```python
# UNIFIED 2018

employment_unified_2018 = pd.read_csv(employment_unified_2018_data)
# pd.DataFrame(employment_unified_2018)
```

```python
# ELEMENTARY 2018

employment_elementary_2018 = pd.read_csv(employment_elementary_2018_data)
# pd.DataFrame(employment_elementary_2018)
```

```python
# SECONDARY 2018 

employment_secondary_2018 = pd.read_csv(employment_secondary_2018_data)
# pd.DataFrame(employment_secondary_2018)
```

```python
list_of_files = [
    "Merged Employment/Resources/output_employment_data/Unified2018.csv",
    "Merged Employment/Resources/output_employment_data/Elementary2018.csv",
    "Merged Employment/Resources/output_employment_data/Secondary2018.csv"
]

# Format pd.read_csv(filename).assign(year=2018)
employment_2018_df = []
for i,filename in enumerate(list_of_files):
    X1 = pd.read_csv(filename)
    X1["year"] = 2018
    employment_2018_df.append(X1)
    
# Combine the data into a single dataset
employment_2018_data = pd.concat(employment_2018_df)
employment_2018_merged = pd.DataFrame(employment_2018_data)
```

```python
# Save csv file 

employment_2018_merged.to_csv("Merged Employment/Resources/merged_employment_yearly/Merged_Employment_2018.csv")
```
